Stratistical Operations-

```python
print(np.mean(arrx))
print(arrx.sum())
print(np.median(arry)) 
print(np.var(arry)) 
print(np.min(arry)) 
print(np.max(arry)) 
print(np.percentile(arry,40)) 
print(np.sort(arry)) 
```

Copy and View-
Copy does not change the original data, but View does.

```python
arr_copy = arr1.copy()
arr_view = arr1.view()

arr_copy[0]= 40
print(arr1)
print(arr_copy)
```

```python
arr_view[2]= 500
print(arr1)
print(arr_view)
```
