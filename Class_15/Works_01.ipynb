```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y)
plt.title('Square Function')
plt.show()
```

Marker Property
We can use data highlight to point specific points in data.


```python
heroes = ['Prabhas', 'Pawan', 'Chiranjeevi', 'Sharukh', 'Amitabh'] #x-axis values



movies = [100,300,200,600,1000] #height of bars, valuesfor y-axis

c = ['r', 'b','k','g', 'orange']

plt.bar (heroes, movies, color=c)

plt.xlabel('Hero Name', color='b', fontsize=15)

plt.ylabel('Number of Movies', color='b', fontsize=15)

plt.title('Hero wise number of movies',color='r', fontsize=15)

plt.show()
```

```python
Classes = ['5th', '6th', '7th','8th','9th','10th']
Results = [90,89,99,88,94,97]
plt.bar(Classes,Results,color='r',width=0.05)
plt.plot(Classes,Results,color='b')
# plt.hist(Classes,Results,color='b')
plt.xlabel('Classes',color='orange',fontsize=15)
plt.ylabel('Results',color='b',fontsize=12)
plt.title('Results of Classes',color='g',fontsize=20)
plt.show()
```

```python

a = np.arange(10)

plt.grid();
plt.plot(a,a, marker='o', label='identity')

plt.plot(a,a**2, marker='o', label='square')

plt.plot(a,a**3, marker='o', label='cubic')

plt.legend()

plt.show()
```

PIE CHART

```python
marks = np.array([45, 88, 75, 90, 60])
mylabels = ['Math', 'Science', 'History', 'English', 'Physics']   
c = ['r', 'b', 'g','y','c','m','k','w'] 
plt.pie(marks, labels = mylabels, colors = c)
plt.show()
```

```python

marks = np.array([45, 88, 75, 90, 60])
mylabels = ['Math', 'Science', 'History', 'English', 'Physics']
myexplode = [0.1, 0.2, 0, 0, 0]
plt.pie(marks,labels = mylabels, explode = myexplode,autopct = '%1.1f%%')
plt.show()
```
